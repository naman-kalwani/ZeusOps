```python
# sanity_check.py
import numpy as np, pandas as pd
from pathlib import Path

DATA_DIR = Path("/content/drive/MyDrive/zeusOps/data/UNSW-NB15")
x_train = pd.read_pickle(DATA_DIR / "unsw_x_train_v2.pkl")
x_test  = pd.read_pickle(DATA_DIR / "unsw_x_test_v2.pkl")
y_train = pd.read_pickle(DATA_DIR / "unsw_y_train_v2.pkl")
y_test  = pd.read_pickle(DATA_DIR / "unsw_y_test_v2.pkl")

# make y one-dimensional numpy arrays
y_train = np.asarray(y_train).ravel()
y_test  = np.asarray(y_test).ravel()

print("Train counts:\n", pd.Series(y_train).value_counts().sort_index())
print("Test counts:\n",  pd.Series(y_test).value_counts().sort_index())

# check the unique labels
print("Unique labels:", np.unique(y_train))
# confirm normal label location (you used 7 as normal earlier)
print("Proportion normal in train:", np.mean(y_train==7))
print("Proportion normal in test:",  np.mean(y_test==7))
```
